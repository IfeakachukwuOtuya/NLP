## Text to Speech conversion using GTTS

```python
#!pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio
text_to_speech = gTTS('''my name is Ifeakachukwu Otuya and am a data scientist and AI, with experience across 
                        industries like Finance, Engineering, Health care, Education and operational. with skills including Python
                        SQL, Machine Learning, NLP, Deeplearning, GenerativeAI and Agentic AI ''')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay=False)
    
```

## Text to Speedch conversion using pyttsx3

```python
#!pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text = '''my name is ife and am a data scientist and AI, with experience across 
        industries like Finance, Engineering, Health care, Education and operational. with skills including Python
        SQL, Machine Learning, NLP, Deeplearning, GenerativeAI and Agentic AI '''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

#change the vioce
voice = audio.getProperty('voices')

# 0 for male ad 1 for female
audio.setProperty('voice', voice[0].id)    # for male voice

#audio.setProperty('voice', voice[1].id)    # for female voice

# text-to-speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'text_male_Vioce.mp3')
#audio.save_to_file(text, 'text_female_Vioce.mp3')

audio.runAndWait()

```
