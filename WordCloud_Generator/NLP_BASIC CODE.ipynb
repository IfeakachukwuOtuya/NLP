```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = 'Diabetes is a chronic medical condition that occurs when the body cannot effectively regulate blood sugar (glucose) levels. There are two main types: type 1 diabetes, an autoimmune disease where the pancreas produces little or no insulin, and type 2 diabetes, which is linked to insulin resistance and often associated with obesity, poor diet, and lack of physical activity. Common symptoms include excessive thirst, frequent urination, fatigue, and blurred vision. Long-term complications can include nerve damage (neuropathy), kidney disease, cardiovascular problems, and eye disorders like retinopathy. Management of diabetes typically involves regular monitoring of blood glucose, insulin therapy, healthy eating, exercise, and medication. Early diagnosis and lifestyle changes can significantly reduce the risk of complications.'
text
```

```python
wordcloud = WordCloud(width=460, height=250, margin=3, background_color='white', colormap='Accent', mode='RGBA').generate(text)
```

```python
plt.imshow(wordcloud, interpolation='quadric')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```

Embedding --- Transformer every text will convert to number

```python
from sklearn.feature_extraction.text import CountVectorizer

sentence = ' data science and ai genai has great career ahead data science'
vectorizer = CountVectorizer()
vector = vectorizer.fit_transform([sentence])
vector.toarray()

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
sentence = ' data science and ai genai has great career ahead data science'
vectorizer = TfidfVectorizer()
vector = vectorizer.fit_transform([sentence])
vector.toarray()
```
